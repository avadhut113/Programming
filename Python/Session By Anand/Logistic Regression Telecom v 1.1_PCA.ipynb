```python
# Predicted probabilities
y_pred = logsk.predict_proba(X_test[col])
```

### Model Evaluation

```python
# Confusion matrix 
confusion = metrics.confusion_matrix( y_pred_final.Churn, y_pred_final.predicted )
confusion
```

```python
TP = confusion[1,1] # true positive 
TN = confusion[0,0] # true negatives
FP = confusion[0,1] # false positives
FN = confusion[1,0] # false negatives
```

```python
# Let's see the sensitivity of our logistic regression model : Positive, with diseasee
# True positive rate
TP / float(TP+FN)
```

```python
# Let us calculate specificity
TN / float(TN+FP)
```

```python
# Calculate false postive rate - predicting churn when customer does not have churned
FP/ float(TN+FP)
```

```python
# Calculate false negative rate - predicting not churn when customer have churned

FN/ float(FN+TP)
```

```python
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(6, 4))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return fpr, tpr, thresholds
```

```python
draw_roc(y_pred_final.Churn, y_pred_final.predicted)
```

```python
pca.components_
```

```python
colnames = list(X_train.columns)
```

```python
pcs_df = pd.DataFrame({'PCompA':pca.components_[0],'PCompB':pca.components_[1], 'Features':colnames})
```

```python
pcs_df.head(10)
```

```python
# Plot Principlal Components
fig = plt.figure(figsize = (8,8))
plt.scatter(pcs_df.PCompA, pcs_df.PCompB)
plt.xlabel('Principal Component A')
plt.ylabel('Principal Component B')
for i, txt in enumerate(pcs_df.Features):
    plt.annotate(txt, (pcs_df.PCompA[i],pcs_df.PCompB[i]))
plt.tight_layout()
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
#Making the screeplot - plotting the cumulative variance against the number of components
%matplotlib inline
fig = plt.figure(figsize = (10,10))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('No of components')
plt.ylabel('Variance')
plt.show()
```

```python
from sklearn.decomposition import IncrementalPCA
pca_final = IncrementalPCA(n_components=16)
```

```python
df_train_pca = pca_final.fit_transform(X_train)
df_train_pca.shape
```

```python
#creating correlation matrix for the principal components
corrmat = np.corrcoef(df_train_pca.transpose())
```

```python
#plotting the correlation matrix
%matplotlib inline
plt.figure(figsize = (15,10))
sns.heatmap(corrmat,annot = True)
```

```python
corrmat_nodiag = corrmat - np.diagflat(corrmat.diagonal())
print("max corr:",corrmat_nodiag.max(), ", min corr: ", corrmat_nodiag.min(),)
```

```python
df_test_pca = pca_final.transform(X_test)
df_test_pca.shape
```

```python
#Training the model on the train data
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

learner_pca = LogisticRegression()
model_pca = learner_pca.fit(df_train_pca,y_train)
```

```python
#Making prediction on the test data
pred_probs_test = model_pca.predict_proba(df_test_pca)[:,1]
"{:2.2}".format(metrics.roc_auc_score(y_test, pred_probs_test))
```

```python
%matplotlib inline
fig = plt.figure(figsize = (8,8))
plt.scatter(df_train_pca[:,0], df_train_pca[:,1], c = y_train.map({0:'green',1:'red'}))
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.tight_layout()
plt.show()
```
